```python
#Shape hierarchy with ABC
from abc import ABC, abstractmethod
import math

class Shape(ABC):

    count = 0

    def __init__(self):
        Shape.count += 1

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def describe(self):
        print("Shape :", self.__class__.__name__)
        print("Area =", self.area())
        print("Perimeter =", self.perimeter())
        print()

    @classmethod
    def total_shapes(cls):
        return cls.count


class Circle(Shape):

    def __init__(self, radius):
        super().__init__()
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius


class Rectangle(Shape):

    def __init__(self, length, width):
        super().__init__()
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


class Triangle(Shape):

    def __init__(self, a, b, c):
        super().__init__()
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))


def print_report(shapes_list):
    for shape in shapes_list:
        shape.describe()


c = Circle(5)
r = Rectangle(6, 4)
t = Triangle(3, 4, 5)

shapes = [c, r, t]

print_report(shapes)

print("Total Shapes =", Shape.total_shapes())
```

```python
#Vehicle fleet system
class Vehicle:

    count = 0

    def __init__(self, name):
        self.name = name
        Vehicle.count += 1

    def fuel_cost(self, km):
        pass


class Car(Vehicle):

    def __init__(self, name):
        super().__init__(name)

    def fuel_cost(self, km):
        return km * 8


class Truck(Vehicle):

    def __init__(self, name):
        super().__init__(name)

    def fuel_cost(self, km):
        return km * 15


class Motorcycle(Vehicle):

    def __init__(self, name):
        super().__init__(name)

    def fuel_cost(self, km):
        return km * 4


class ElectricCar(Car):

    def __init__(self, name):
        super().__init__(name)

    def fuel_cost(self, km):
        return 0


def fleet_report(vehicle_list):
    for vehicle in vehicle_list:
        print("Vehicle :", vehicle.name)
        print("Fuel Cost for 10 km =", vehicle.fuel_cost(10))

        if isinstance(vehicle, ElectricCar):
            print("This is an Electric Car")

        print()


c = Car("Swift")
t = Truck("Tata Truck")
m = Motorcycle("Splendor")
e = ElectricCar("Tesla")

vehicles = [c, t, m, e]

fleet_report(vehicles)

print("Total Vehicles =", Vehicle.count)
```

```python
#Multi Role Staff System
class Person:

    def __init__(self, name):
        self.name = name


class Employee(Person):

    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id


class Teacher(Employee):

    def __init__(self, name, emp_id, subjects):
        Employee.__init__(self, name, emp_id)
        self.subjects = subjects

    def teach(self):
        print(self.name, "is teaching", ", ".join(self.subjects))


class AdminStaff(Employee):

    def __init__(self, name, emp_id, designation):
        Employee.__init__(self, name, emp_id)
        self.designation = designation

    def admin_task(self):
        print(self.name, "is performing admin work as", self.designation)


class TeacherAdmin(Teacher, AdminStaff):

    def __init__(self, name, emp_id, subjects, designation):
        Teacher.__init__(self, name, emp_id, subjects)
        self.designation = designation

    def admin_task(self):
        print(self.name, "is performing admin work as", self.designation)


t = TeacherAdmin(
    "Mayank",
    101,
    ["Python", "AI & ML"],
    "HOD"
)

t.teach()
t.admin_task()

print("\nMRO:")
print(TeacherAdmin.__mro__)
```
