```python
#secure student record 
class Student:
    count_students = 0

    def __init__(self, roll_no, marks, grade):
        self.__roll_no = roll_no
        self._grade = grade
        self.marks = marks

        Student.count_students += 1

    @property
    def marks(self):
        return self.__marks

    @marks.setter
    def marks(self, marks):
        for m in marks:
            if m < 0 or m > 100:
                raise ValueError("Marks should be between 0 and 100")
        self.__marks = marks

    @property
    def gpa(self):
        return sum(self.__marks) / len(self.__marks) / 10

    @classmethod
    def count(cls):
        return cls.count_students


# Testing
s1 = Student(1, [80, 90, 85], "A")
s2 = Student(2, [70, 75, 80], "B")

print("Student 1 GPA:", s1.gpa)
print("Student 2 GPA:", s2.gpa)
print("Total Students:", Student.count())
```

```python
#library book manager
class Book:

    def __init__(self, isbn, title, author, copies):
        self.__isbn = isbn
        self._title = title
        self._author = author
        self.__copies = copies

    @property
    def isbn(self):
        return self.__isbn

    @property
    def available(self):
        return self.__copies

    def checkout(self, n):
        if n > self.__copies:
            raise ValueError("Not enough copies available")
        self.__copies -= n

    def return_book(self, n):
        if n <= 0:
            raise ValueError("Invalid number of books")
        self.__copies += n


# Testing
b1 = Book("978123456", "Python Basics", "John", 10)

print("Available Copies:", b1.available)

b1.checkout(3)
print("After Checkout:", b1.available)

b1.return_book(2)
print("After Return:", b1.available)
```

```python
#atm machine simulation
class ATM:

    def __init__(self, pin, owner, balance=0):
        self.__pin = pin
        self._owner = owner
        self.__balance = balance
        self.__authenticated = False

    def authenticate(self, pin):
        if pin == self.__pin:
            self.__authenticated = True
            print("Login Successful")
        else:
            print("Wrong PIN")

    @property
    def balance(self):
        return self.__balance

    def deposit(self, amount):
        if self.__authenticated:
            self.__balance += amount
            print("Deposited:", amount)
        else:
            print("Please login first")

    def withdraw(self, amount):
        if not self.__authenticated:
            print("Please login first")
            return

        if amount > 20000:
            print("Maximum withdrawal limit is ₹20000")
        elif amount > self.__balance:
            print("Insufficient Balance")
        else:
            self.__balance -= amount
            print("Withdrawn:", amount)

    def mini_statement(self):
        print("Owner:", self._owner)
        print("Current Balance:", self.__balance)


# Testing
a1 = ATM(1234, "Mayank", 50000)

a1.authenticate(1234)
a1.deposit(5000)
a1.withdraw(10000)
a1.mini_statement()
```
