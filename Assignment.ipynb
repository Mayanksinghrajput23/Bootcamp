```python
#Custom Sorter
students = [
    {"name": "Mayank", "gpa": 8.7},
    {"name": "Ritu", "gpa": 9.2},
    {"name": "Himanshi", "gpa": 8.9},
    {"name": "Parth", "gpa": 9.5},
    {"name": "Shivanya", "gpa": 8.5}
]

# Sort by GPA in descending order
sorted_students = sorted(students, key=lambda x: x["gpa"], reverse=True)

# Print top 3 students
print("Top 3 Students:")
for student in sorted_students[:3]:
    print(student)
```

```python
#Data Pipeline
from functools import reduce

numbers = list(range(1, 21))

result = reduce(
    lambda x, y: x + y,
    map(
        lambda x: x ** 2,
        filter(lambda x: x % 2 != 0, numbers)
    )
)

print("Sum of squares of odd numbers:", result)
```

```python
#Mini Calculator Factory
def make_operator(op):
    if op == "+":
        return lambda a, b: a + b
    elif op == "-":
        return lambda a, b: a - b
    elif op == "*":
        return lambda a, b: a * b
    elif op == "/":
        return lambda a, b: a / b
    else:
        return lambda a, b: "Invalid Operator"

# Create functions
add = make_operator("+")
subtract = make_operator("-")
multiply = make_operator("*")
divide = make_operator("/")

# Test with two numbers
a = 23
b = 7

print("Addition =", add(a, b))
print("Subtraction =", subtract(a, b))
print("Multiplication =", multiply(a, b))
print("Division =", divide(a, b))
```

```python
#Safe input validator
class NegativeNumberError(Exception):
    pass

def get_positive_int():
    while True:
        try:
            n = int(input("Enter a positive integer: "))

            if n < 0:
                raise NegativeNumberError("Negative number not allowed")

            return n

        except ValueError:
            print("Please enter a valid integer")

        except NegativeNumberError as e:
            print(e)

num = get_positive_int()
print("Valid Number =", num)
```

```python
#Rebust file processor
import csv

# Create sample CSV file
with open("data.csv", "w") as f:
    f.write("10,20,30\n20,30,40\n30,40,50")

def process_csv(filepath):
    file = None

    try:
        file = open(filepath, "r")
        reader = csv.reader(file)

        data = list(reader)

        rows = len(data)
        cols = len(data[0])

        averages = []

        for j in range(cols):
            total = 0

            for i in range(rows):
                total += float(data[i][j])

            avg = total / rows
            averages.append(avg)

        print("Column Averages:", averages)

    except FileNotFoundError:
        print("Error: File not found")

    except PermissionError:
        print("Error: Permission denied")

    except csv.Error:
        print("Error: Invalid CSV format")

    finally:
        if file:
            file.close()
            print("File closed successfully")

# Function Call
process_csv("data.csv")
```

```python
#Bank Account
class InsufficientFundsError(Exception):
    pass

class InvalidAmountError(Exception):
    pass

class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Invalid deposit amount")
        
        self.balance += amount
        print("Deposited:", amount)
        print("Balance:", self.balance)

    def withdraw(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Invalid withdrawal amount")
        
        if amount > self.balance:
            raise InsufficientFundsError("Insufficient balance")
        
        self.balance -= amount
        print("Withdrawn:", amount)
        print("Balance:", self.balance)

# Testing
account = BankAccount(1000)

transactions = [
    ("deposit", 500),
    ("withdraw", 200),
    ("withdraw", 2000),
    ("deposit", -100),
    ("withdraw", 300)
]

for t_type, amount in transactions:
    try:
        if t_type == "deposit":
            account.deposit(amount)
        elif t_type == "withdraw":
            account.withdraw(amount)

    except InvalidAmountError as e:
        print("Error:", e)

    except InsufficientFundsError as e:
        print("Error:", e)
```
